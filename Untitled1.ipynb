```python
flat=[]
trend=[]
no_data=[]
for i in df['CARD_SIDO_NM'].unique():
    for j in df['STD_CLSS_NM'].unique():
        df_set=df.loc[(df['CARD_SIDO_NM']==i)&(df['STD_CLSS_NM']==j)]
        df_re=df_set.groupby(['REG_YYMM','CARD_SIDO_NM','STD_CLSS_NM'])[['AMT']].sum()
        df_re=df_re.reset_index().reset_index()
        df_re['Month']=df_re['index']+1
        df_re=df_re.drop(['index'],axis=1)
        Y=df_re['AMT']
        X=df_re['Month']
        try:
            X=sm.add_constant(X)
            model = sm.OLS(Y,X)
            results = model.fit()
            if results.pvalues['Month']>0.05:
                  flat.append((i,j))
            else:
                  trend.append((i,j))
        except:
            no_data.append((i,j))
```

```python
# 예측 템플릿 만들기
year = 202000
months  = [4,5,6,7]
CARD_SIDO_NMs=sorted(df['CARD_SIDO_NM'].unique())
STD_CLSS_NMs=sorted(df['STD_CLSS_NM'].unique())
temp = []
for month in months:
    for i in CARD_SIDO_NMs:
        for j in STD_CLSS_NMs:
            temp.append([year+month,i,j,0])
temp = np.array(temp)
temp = pd.DataFrame(data=temp, columns=['REG_YYMM','CARD_SIDO_NM','STD_CLSS_NM','AMT'])
```
